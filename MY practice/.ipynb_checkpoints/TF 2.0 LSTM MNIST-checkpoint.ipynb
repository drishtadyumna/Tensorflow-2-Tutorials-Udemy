```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
i = tf.keras.layers.Input(shape=(X_train[0].shape))
x = tf.keras.layers.LSTM(128)(i)
x = tf.keras.layers.Dense(10, activation='softmax')(x)

model = tf.keras.models.Model(i, x)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
r = model.fit(X_train, y_train, validation_data=[X_test, y_test], epochs=10, batch_size=500)
```

```python
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```

```python
plt.plot(r.history['accuracy'], label='accuracy')
plt.plot(r.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test, verbose='quiet')
```

```python
y_hat = np.argmax(model.predict(X_test), axis=1)
indices = np.where(y_hat != y_test)[0]
i = list(np.random.choice(indices, 3))
for index in i:
    plt.imshow(X_test[index], cmap='gray')
    plt.title('Predicted: {}, Actual: {}'.format(y_hat[index], y_test[index]))
    plt.show()
```
